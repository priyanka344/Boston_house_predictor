```python
sns.distplot(data['salesprice']);
```

A distplot helps us to plot the variable agaisnt the density variation


```python
print("skewness: %f" % data['salesprice'].skew())
print("kurtosis: %f" % data['salesprice']. kurt())
```

skewness is the measure of asymmetry in the distribution

kurtosis is the measure of how often the outliers occur in the given graph. With medium level kurtosis we have mesokurtic and low level kurtosis we have platykurotic

```python
fig, ax=plt.subplots()
ax.scatter(x=data['CRIM'], y=data['salesprice'])
plt.ylabel('salesprice', fontsize=13)
plt.xlabel('CRIM', fontsize=13)
plt.show()
```

```python
from scipy import stats
from scipy.stats import norm, skew #for some statistics
sns.distplot(data['salesprice'], fit=norm);
(mu, sigma)=norm.fit(data['salesprice'])
print('\n mu={:.2f} and sigma={:.2f}\n'.format(mu, sigma))
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=${:.2f})', format(mu, sigma)], loc='best')
plt.ylabel('frequency')
plt.title('salesprice distribution')

#get also the qq plot
fig=plt.figure()
res=stats.probplot(data['salesprice'], plot=plt)
plt.show()

```

scipy provides utility function for optimization, stats and signal processing.

norm is normal continous random variable



```python
data["salesprice"]=np.log1p(data["salesprice"])      #we use the log function to avoid skewness or outliers
sns.distplot(data['salesprice'], fit=norm);
(mu, sigma)=norm.fit(data['salesprice'])
print('\n mu={:.2f} and sigma={:.2f}\n'.format(mu,sigma))
plt.legend({'Normal dist.($\mu=${:.2f} and $\sigma=${:.2f}'.format(mu, sigma)}, loc='best')
plt.ylabel('frequency')
plt.title('salesprice distribution')
fig=plt.figure()
res=stats.probplot(data['salesprice'], plot=plt)
plt.show()
```

##Data correlation


Data correlation is the relationship between different features. +1 is for positive relationship and -1 is for negative relationship

```python
plt.figure(figsize=(10,10))
cor=data.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.PuBu)
plt.show()
```

Heatmaps are used to visualize the content better and direct the users towards areas which matter most

```python
cor_target=abs(cor['salesprice']) #absolute value of the correlation
relevant_features=cor_target[cor_target>0.2] #highly correlated features
names=[index for index, value in relevant_features.iteritems()]   #getting the names of the features
names.remove('salesprice')  #removing target features
print(names)
print(len(names))
```

## Model Building

```python
from sklearn.model_selection import train_test_split
x=data.drop("salesprice", axis=1)
y=data["salesprice"]
x_train, x_test, y_train, y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
```

```python
predictions=lr.predict(x_test)
print("actual value of the house:- ", y_test[0])
print("model predicted value: ", predictions[0])
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test, predictions)
rmse=np.sqrt(mse)
print(mse)
```
